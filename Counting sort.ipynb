```python
def countingSort(array):
    # stable counting sort
    
    size = len(array)
    
    output = [0] * size
    
    count = [0] * size
    
    for i in range(size):
        
        count[array[i]] += 1
        
    for i in range(1, size):
        
        count[i] += count[i-1]
    
    for i in range((size-1), -1, -1):           
        
        output[count[array[i]] - 1] = array[i]
        
        count[array[i]] -= 1
        
        i -= 1
        
    return output

print(countingSort([4,1,3,2,3]))
#print(countingSort([4,1,5,2,3]))
```
